```python
     
def grad_des(X_test1,Y_test):
    m=0
    c=0
    learning_rate=0.01
    n=len(X_test)
    for i in range(600):
        y_predictions=m*X_test+c
        #now doing cost function
        cost_fun=(1/n)*sum([j**2 for j in (Y_test-y_predictions)])
        #drivatives
        d_slope=-(2/n)*sum(-X_test1*(Y_test-(y_predictions)))
        #derivative for constat
        d_constant=(2/n)*sum([ Y_test-y_predictions])
        acc=r2_score(Y_test,y_predictions)
        print('slope :',m,'intercept :',c,'cost function :',cost_fun,'Acuuracy :',acc)

        
        #step size
        
        step_size_m=learning_rate*d_slope
        step_size_c=learning_rate*d_constant
        
        
         
        m=m+ step_size_m
        c=c+step_size_m
        
```

```python
X_test=X_test.flatten()
grad_des(X_test,Y_test)
```
